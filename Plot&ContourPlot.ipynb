```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x1,x2):
    return  (x1 - x2)**4 + 8*x1*x2 - x1 + x2 + 3
```

```python
x1 = np.linspace(-5,5,100)
x2 = [-1,1]
```

```python
y = function(x1,x2[0])
plt.plot(x1, y, color='red', label=f'x2 = {x2[0]}')

y = function(x1,x2[1])
plt.plot(x1, y, color='green', label=f'x2 = {x2[0]}')

plt.xlabel('x1')
plt.ylabel('f(x1, x2)')
plt.title('Plot of f(x1, x2) for different x2 values')
plt.legend()
plt.show()
```
