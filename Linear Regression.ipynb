```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x  = np.array([1,3,5]) # Used for years of experience 
y = np.array([300,480,570]) # Salary
print(x)
print(y)
```

```python
plt.plot(x, y, color='red')
plt.xlabel('Experience(years)')
plt.ylabel('Salary')
plt.title('Plot of Salary and years experience')
plt.legend()
plt.show()
```

```python
def forward(x,w,b):
    return w*x + b
```

```python
# LOSS = mean square error 
def compute_cost(y,y_hat):
    return ((y - y_hat)**2).mean()
```

```python
def compute_gradient(y,y_hat):
    dw =  np.dot(2*x,y_hat - y).mean()
    db = (2*(y_hat - y)).mean()
    return dw,db
```

```python
# Back propagation 
alpha = 0.01
itr = 10000

def compute_gradient_descent(x, y,w,b,alpha, itr):
    for i in range(itr):
        # Forward propagation
        y_hat = forward(x,w,b)
        
        # loss function
        l = compute_cost(y,y_hat)
        
        # Gradients 
        dw , db = compute_gradient(y,y_hat)
        
        w = w - alpha*dw 
        b = b - alpha*db
        
    return w,b
```

```python
alpha = 0.01
itr = 10000
w,b = compute_gradient_descent(x, y,0,0, alpha, itr)
print(f"The salary after {7} years of experience is {forward(7,w,b)}")
```
